```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load cleaned dataset
df = pd.read_csv("data/cleaned_sales.csv")

# Convert 'Date' column to datetime
df["Date"] = pd.to_datetime(df["Date"])

# Display basic information
print("Dataset Overview:")
df.info()
```

```python
# Display first few rows
display(df.head())

# Summary statistics
display(df.describe())
```

```python
# Category distribution
plt.figure(figsize=(8, 5))
sns.countplot(y=df["Category"], order=df["Category"].value_counts().index)
plt.title("Product Category Distribution")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()
```

```python
# Sales trend over time
plt.figure(figsize=(12, 6))
sns.lineplot(x="Date", y="Revenue", data=df, estimator='sum')
plt.title("Sales Revenue Over Time")
plt.xlabel("Date")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python


# Plot monthly sales
plt.figure(figsize=(12, 6))
sns.boxplot(x='Month', y='Revenue', data=df)
plt.title('Monthly Sales Distribution')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.show()

# Plot daily sales
plt.figure(figsize=(12, 6))
sns.boxplot(x='DayOfWeek', y='Revenue', data=df)
plt.title('Daily Sales Distribution')
plt.xlabel('Day of Week')
plt.ylabel('Revenue')
plt.show()
```

```python
# Plot sales on holidays vs non-holidays
plt.figure(figsize=(8, 5))
sns.boxplot(x='Holiday', y='Revenue', data=df)
plt.title('Sales on Holidays vs Non-Holidays')
plt.xlabel('Holiday')
plt.ylabel('Revenue')
plt.show()

# Plot sales during promotions
plt.figure(figsize=(8, 5))
sns.boxplot(x='Promotion_Event', y='Revenue', data=df)
plt.title('Sales During Promotions')
plt.xlabel('Promotion_Event')
plt.ylabel('Revenue')
plt.show()
```

```python
# Correlation matrix
corr_matrix = df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Boxplot for sales
plt.figure(figsize=(8, 5))
sns.boxplot(y='Revenue', data=df)
plt.title('Sales Distribution')
plt.show()

```
